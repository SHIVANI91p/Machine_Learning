# Model Building

```python
import statsmodels.formula.api as smf  # we use statsmodels library to perform hypothesis testing H0: M(SLOP)=0 colmun is not important .Ha: M(SLOP)!=0  column is importantt i.e HP & SP colmns are important  
model = smf.ols('MPG~HP+VOL+SP+WT',data=df).fit() # fit() is use for calculation     by Taking WT & VOL columns together cuasing multicollinearity problem
```

```python
model.pvalues
```

# Linear Regression

```python
model = smf.ols('MPG~HP+VOL+SP+WT',data=df).fit() # by Taking independent WT not cuasing multicollinearity problem                   
wt = smf.ols('MPG~WT+HP+SP',data = df).fit()
wt.pvalues
```

```python
#vol          # by Taking independent VOL not cuasing multicollinearity problem
vol = smf.ols('MPG~VOL+SP+HP',data = df).fit()
vol.pvalues
```

## 1.Residual Analysis

As per the subset selection, it is clear that VOL column is more important than WT column as it's R2 value is more and AIC value is less,So we will select VOL columns and will drop WT column.

NORMALITY TEST(Q-Q plot)

```python
model = smf.ols('MPG~HP+VOL+SP', data=df).fit()
model.rsquared
```

```python
#Statsmodel
import statsmodels.api as sm      # to plot Q-Q Plot
```

```python
df.head()
```

```python
model.fittedvalues         # predicted values
```

```python
model.resid           # error values
```

```python
#Q-Q plot      quantile-quantile plot which is only use to check normality of data set
sm.qqplot(model.resid, line = 'q'); # for errore values we plot this plot  *actual value vs predicted values=residules(error)
# the data is not normalized
```

```python
#get index for higher residuals
np.argmax(model.resid), np.max(model.resid)    # it gives index number of extreme value
```

# MODEL DELETION TECHNIQUES


# Cook's Distance

* Detecting influencers and outliers

```python
model
```

```python
#Find the influence data
inf = model.get_influence()

#Calculate the cooks distance
c,p = inf.cooks_distance
```

```python
c
```

```python
# Cook's distance plot
plt.stem(c)
plt.xlabel('Index')
plt.ylabel('Cooks Distance')
```

```python
np.argmax(c), np.max(c)    # Cooks Distance (CD)>1 is influencer
```

```python
df.iloc[[76]]
```

```python
df.head()
```

```python
#Influence Plot
from statsmodels.graphics.regressionplots import influence_plot
```

```python
influence_plot(model);
```

```python
#Calculate cutoff
k = len(df.columns)
n = len(df)

lv = 3*(k+1)/n
lv
```

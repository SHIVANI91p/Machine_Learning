# Model building

```python
x = df.iloc[:,:-1]
y = df['price']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3, random_state=1)
```

```python
lr = LinearRegression()
lr.fit(xtrain,ytrain)
ypred = lr.predict(xtest)
ypred
```

```python
lr.intercept_ # incercept(c)
```

```python
lr.coef_  # slop/weight (m)
```

```python
lr.score(xtrain,ytrain), lr.score(xtest,ytest)  # Accuracy
```

# Lasso

```python
l1 = Lasso()
l1.fit(xtrain,ytrain)
l1.score(xtrain,ytrain), l1.score(xtest,ytest)
```

```python
l1.coef_
```

# Hyperparameter Tuning

```python
for i in range(100,200):
    l1 = Lasso(alpha=i)
    l1.fit(xtrain,ytrain)
    print(f"Alpha: {i} Train : {l1.score(xtrain,ytrain)} Test: {l1.score(xtest,ytest)}")
```

```python
l1 = Lasso(alpha=175)
l1.fit(xtrain,ytrain)
l1.score(xtrain,ytrain), l1.score(xtest,ytest)
```

```python
l1.coef_    # so there are two columns which we can remove so this is calles feature selection
```

```python
xtrain.columns  # so we can remove  normalized-losses,fuel-type or else these columns are no that much important
```

# Ridge 

```python
l2 = Ridge()

l2.fit(xtrain,ytrain)
l2.score(xtrain,ytrain), l2.score(xtest,ytest)
```

```python
for i in range(1,100):
    l2 = Ridge(alpha=i)
    l2.fit(xtrain,ytrain)
    print(f"Alpha: {i} Train: {l2.score(xtrain,ytrain)} Test: {l2.score(xtest,ytest)}")
```

```python
l2 = Ridge(alpha=7)   # within less aplha value we got better accuracy so we can select ridge algorithm
l2.fit(xtrain,ytrain)
l2.score(xtrain,ytrain),l2.score(xtest,ytest)
```
